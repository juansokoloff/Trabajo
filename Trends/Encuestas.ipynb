```python
col = []

for i in range (0, len(df)):
    if df.Encuestadora[i] == "Datexco":
        col.append("r")
    elif df.Encuestadora[i] == "Cifras y Conceptos":
        col.append("b")
    elif df.Encuestadora[i] == "Yanhaas":
        col.append("g")
    elif df.Encuestadora[i] == "Invamer":
        col.append("pink")
    elif df.Encuestadora[i] == "Guarumo":
        col.append("orange")
    elif df.Encuestadora[i] == "CNC":
        col.append("grey")
    else:
        col.append("k")
```
